# MatPlotLib



```python
dados.head(5)
```

```python
# Gráfico gerado a partir de coordenadas e colunas
plt.bar(dados['Nome'], dados['Periodo'])
```

```python
plt.plot(dados['Nome'], dados['Idade'], marker= "o")
```

```python
# Agrupar alunos por curso usando pandas
dados_2 = dados.groupby('Curso')[['Nome']].count()
dados_2
```

```python
dados_2.plot(kind = 'bar')
```

```python
# dados em forma de barra
dados_3 = dados.groupby('Sexo')[['Idade']].count()
dados_3
dados_3.plot(kind = 'bar')
```

```python
# Dados no groupby formato de torta ou pizza
dados_4 = dados.groupby('Sexo')[['Nome']].count()
dados_4.plot(kind = 'pie', subplots = True)
```

```python
#Personalização de Dados
dados_4.plot(kind = 'barh', color ='g')
```
