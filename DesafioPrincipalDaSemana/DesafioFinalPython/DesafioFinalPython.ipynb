#Agrupamento de Dados e Plots
- Departamento x Funcionários > plot bar
- Escolaridade x Funcionários > plot line
- Atrito x Funcionário > plot pie

```python
dados_1 = dados.groupby('Department')[['EmployeeNumber']].count()
dados_1.plot(kind = 'bar', color = 'g')
```

Coloquei o dados_2 e o dados_3 para printar, só pra eu visualizar os dados numéricos

```python
dados_2 = dados.groupby('Education')[['EmployeeNumber']].count()
dados_2.plot(kind = 'line', marker = 'o')
dados_2
```

```python
dados_3 = dados.groupby('Attrition')[['EmployeeNumber']].count()
dados_3.plot(kind = 'pie', subplots = True)
dados_3
```
